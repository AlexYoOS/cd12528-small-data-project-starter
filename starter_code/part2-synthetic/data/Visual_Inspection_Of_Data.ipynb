### Inspecting generated synthetic data

```python
# Threshold was set to 0.5 for float to categorical conversion of the Loan Status Column
len(fake[fake['Loan Status'] >= 0.5])
```

```python
# Percentage generated Loan Status == 1 rows
fake['Loan Status'].mean()
```

**Observation:** Expected was a generation of data which columns are only **floats**, and therefore this data differs from the imbalanced part of the original dataset. The set threshold of 0.5 for categorical conversion of the target column **Loan Status**, led to 77% Loan Status == 1 rows. Shifting this threshold would result in different percentages and influences the balancing attempt of the dataset.

Further the synthetic data could/should be made more realistic by manually converting other features back to categorical data, amongst other measures that can be taken to try to optimise the process.
